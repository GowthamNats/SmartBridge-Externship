```python
from transformers import TFT5Model, T5Tokenizer, TFT5ForConditionalGeneration, Seq2SeqTrainingArguments, Seq2SeqTrainer, DataCollatorForSeq2Seq
```

```python
from transformers import AutoModelForSeq2SeqLM

model = AutoModelForSeq2SeqLM.from_pretrained('t5-small')
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('t5-small')
```

```python
max_input_length = 1024
max_target_length = 1024

def preprocess_function(examples):
    inputs = examples[0]
    model_inputs = tokenizer(inputs, max_length=max_input_length, truncation=True)

    # Setup the tokenizer for targets
    labels = tokenizer(text_target=examples[1], max_length=max_target_length, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
tokenized_train_set = train_set.applymap(preprocess_function)
```

```python
tokenized_test_set = test_set.applymap(preprocess_function)
```

```python
tokenized_train_set.loc[0]
```

```python
batch_size = 2400
model_name = "t5-small".split("/")[-1]
args = Seq2SeqTrainingArguments(
    f"{model_name}-finetuned-xsum",
    evaluation_strategy = "epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=1,
    predict_with_generate=True,
    fp16=False
)
```

```python
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

```python
trainer = Seq2SeqTrainer(
    model,
    args,
    train_dataset=tokenized_train_set,
    #eval_dataset=tokenized_test_set,
    data_collator=data_collator,
    tokenizer=tokenizer,
    #compute_metrics=compute_metrics
)
```

```python
trainer.train()
```
