# TASK - 1
Create a pandas dataframe (DataFrame name as 'df') with numpy random values (4 features and 4 observation)

```python
import numpy as np
import pandas as pd
chars = np.array(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])

f1 = np.random.choice(chars, size=4)
f2 = np.random.randint(0, 20, size=4)
f3 = np.random.rand(4)
f4 = np.random.rand(4)
```

```python
# Adding null values
# f1[2] = np.nan cannot add as it is string
# f2[1] = np.nan cannot add as f2 is of type int
f3[0] = np.nan
f4[3] = np.nan
```

```python
#Making into a data frame
info = {'f1': f1, 'f2': f2, 'f3': f3, 'f4': f4}
df = pd.DataFrame(info)

df
```
